```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "dataset_billionaire_index.csv"
df = pd.read_csv(file_path)

# 1. Summary Statistics
# Assuming the dataset has a 'Net Worth' column
print("Summary Statistics for Net Worth:")
print(df['net_worth'].describe())
```

```python
# 2. Top Indian Billionaires (assuming there are 'Country' and 'Net Worth' columns)
top_indian_billionaires = df[df['country'] == 'India'].nlargest(10, 'net_worth')

plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='net_worth', y='name', data=top_indian_billionaires)
plt.title('Top 10 Indian Billionaires by Net Worth in Dollars')
plt.xlabel('Net Worth (in billions)')
plt.ylabel('Billionaire Name')

# Annotate each bar with its value
for index, value in enumerate(top_indian_billionaires['net_worth']):
    plt.text(value, index, f'{value:.2f}', va='center')  # Display value on the bar (two decimal places)

# Add tick markers on the x-axis (customize tick intervals based on your data)
plt.xticks(ticks=range(0, int(top_indian_billionaires['net_worth'].max()) + 10, 10))  # Adjust the step based on your data

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your DataFrame with a 'country' column

# 3. Country Representation (Top 5 countries by count of billionaires)
top_countries = df['country'].value_counts().head(5)

plt.figure(figsize=(10, 6))

# Set a color list with the first four bars being a default color and the highest bar being highlighted
colors = ['lightgray'] * 5  # Default color for all bars
max_index = top_countries.idxmax()  # Get the country with the highest number of billionaires
colors[list(top_countries.index).index(max_index)] = 'orange'  # Highlight the bar for the country with the max count

# Create the bar plot with specified colors
barplot = sns.barplot(x=top_countries.index, y=top_countries.values, palette=colors)
plt.title('Top 5 Countries by Billionaire Representation')
plt.xlabel('Country')
plt.ylabel('Number of Billionaires')

# Annotate each bar with the number of billionaires
for index, value in enumerate(top_countries.values):
    plt.text(index, value, f'{value}', ha='center', va='bottom')  # Display value above the bar

plt.show()
```

```python
# 4. Industry Representation (Top 5 industries with the most billionaires)
top_industries = df['industry'].value_counts().head(5)

plt.figure(figsize=(10, 6))
barplot = sns.barplot(x=top_industries.index, y=top_industries.values)
plt.title('Top 5 Industries by Billionaire Representation')
plt.xlabel('Industry')
plt.ylabel('Number of Billionaires')
plt.xticks(rotation=45)

# Annotate each bar with the number of billionaires
for index, value in enumerate(top_industries.values):
    plt.text(index, value, f'{value}', ha='center', va='bottom')  # Display value above the bar

plt.show()
```
