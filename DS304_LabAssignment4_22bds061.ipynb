```
#Importing the dataset and Creating a new column 'Zone' and categorizing each institute
import pandas as pd

# Load your dataset
nirf_data = pd.read_csv('NIRF_Overall_2024.csv')

# Create a mapping for the zones
zone_mapping = {
    'Northern Zone': ['Chandigarh', 'Delhi', 'Haryana', 'Himachal Pradesh', 'Jammu and Kashmir', 'Ladakh', 'Punjab', 'Rajasthan'],
    'North Eastern Zone': ['Assam', 'Arunachal Pradesh', 'Manipur', 'Meghalaya', 'Mizoram', 'Nagaland', 'Sikkim', 'Tripura'],
    'Central Zone': ['Chhattisgarh', 'Madhya Pradesh', 'Uttarakhand', 'Uttar Pradesh'],
    'Eastern Zone': ['Bihar', 'Jharkhand', 'Odisha', 'West Bengal'],
    'Western Zone': ['Dadra and Nagar Haveli and Daman and Diu', 'Goa', 'Gujarat', 'Maharashtra'],
    'Southern Zone': ['Andhra Pradesh', 'Karnataka', 'Kerala', 'Puducherry', 'Tamil Nadu', 'Telangana', 'Andaman and Nicobar']
}

# Function to map each state to a zone
def get_zone(state):
    for zone, states in zone_mapping.items():
        if state in states:
            return zone
    return 'Other'

# Apply the function to create the new column
nirf_data['Zone / Zonal Region'] = nirf_data['State'].apply(get_zone)

# Display the first few rows to check the new column
nirf_data.head()
```

```
#Top 5 States with the Highest Number of Ranked Institutions

import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

# Load the dataset
file_path = 'NIRF_Overall_2024.csv'
nirf_data = pd.read_csv(file_path)

# Count the number of institutions per state
state_counts = nirf_data['State'].value_counts()

# Get the top 5 states with the highest number of institutions
top_5_states = state_counts.nlargest(5)

# Plot the top 5 states with the highest number of ranked institutions
plt.figure(figsize=(10, 6))
ax = top_5_states.plot(kind='barh', color='teal')
plt.title('Top 5 States with the Highest Number of Ranked Institutions')
plt.xlabel('Number of Institutions')
plt.ylabel('State')

# Set x-axis to show whole numbers
ax.xaxis.set_major_locator(MaxNLocator(integer=True))

plt.show()

```

```
#3.Zone with the Highest Representation in NIRF Rankings
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
# Replace the path with the actual path to your dataset
file_path = 'NIRF_Overall_2024.csv'
nirf_data = pd.read_csv(file_path)

# Define zones and their states
zones = {
    'Northern Zone': ['Chandigarh', 'Delhi', 'Haryana', 'Himachal Pradesh', 'Jammu and Kashmir', 'Ladakh', 'Punjab', 'Rajasthan'],
    'North Eastern Zone': ['Assam', 'Arunachal Pradesh', 'Manipur', 'Meghalaya', 'Mizoram', 'Nagaland', 'Sikkim', 'Tripura'],
    'Central Zone': ['Chhattisgarh', 'Madhya Pradesh', 'Uttarakhand', 'Uttar Pradesh'],
    'Eastern Zone': ['Bihar', 'Jharkhand', 'Odisha', 'West Bengal'],
    'Western Zone': ['Dadra and Nagar Haveli and Daman and Diu', 'Goa', 'Gujarat', 'Maharashtra'],
    'Southern Zone': ['Andhra Pradesh', 'Karnataka', 'Kerala', 'Puducherry', 'Tamil Nadu', 'Telangana', 'Andaman and Nicobar']
}

# Create a function to categorize states into zones
def categorize_zone(state):
    for zone, states in zones.items():
        if state in states:
            return zone
    return 'Unknown Zone'

# Apply the function to create a new column 'Zone'
nirf_data['Zone'] = nirf_data['State'].apply(categorize_zone)

# Count the number of institutions per zone
zone_counts = nirf_data['Zone'].value_counts()

# Plot the number of institutions per zone
plt.figure(figsize=(10, 6))
zone_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Institutions per Zone in NIRF Rankings')
plt.xlabel('Zone')
plt.ylabel('Number of Institutions')
plt.xticks(rotation=45)
plt.show()

```

```
#Q4. Lie Factor Visualization
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'NIRF_Overall_2024.csv'
nirf_data = pd.read_csv(file_path)

# Count the number of institutions per state
state_counts = nirf_data['State'].value_counts()

# Get the top 5 states with the highest number of institutions
top_5_states = state_counts.nlargest(5)

# Create a figure with two subplots
fig, axs = plt.subplots(1, 2, figsize=(16, 6))

# First plot: Correct Bar Chart with Proper Y-Axis Scale
axs[0].bar(top_5_states.index, top_5_states.values, color='skyblue')
axs[0].set_title('Correct Bar Chart')
axs[0].set_xlabel('State')
axs[0].set_ylabel('Number of Institutions')
axs[0].set_ylim(0, top_5_states.max() + 5)  # Proper y-axis scale
axs[0].set_xticklabels(top_5_states.index, rotation=45)

# Second plot: Misleading Bar Chart with Manipulated Y-Axis Scale
axs[1].bar(top_5_states.index, top_5_states.values, color='lightcoral')
axs[1].set_title('Misleading Bar Chart with Manipulated Y-Axis Scale')
axs[1].set_xlabel('State')
axs[1].set_ylabel('Number of Institutions')
axs[1].set_ylim(top_5_states.min(), top_5_states.max() + 2)  # Compressed y-axis scale
axs[1].set_xticklabels(top_5_states.index, rotation=45)

# Display the plots
plt.tight_layout()
plt.show()


```
