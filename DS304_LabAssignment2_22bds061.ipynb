```
# 2. Count of cars for each cylinder type
plt.figure(figsize=(8, 6))
sns.countplot(x='cyl', data=df, palette='viridis')
plt.title('Count of Cars by Cylinder Type')
plt.xlabel('Number of Cylinders')
plt.ylabel('Count')
plt.show()
```

```
# 3. Proportion of cars with automatic (0) vs manual (1) transmission
transmission_counts = df['am'].value_counts(normalize=True)
transmission_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue'], labels=['Automatic', 'Manual'], startangle=140)
plt.title('Proportion of Cars by Transmission Type')
plt.ylabel('')
plt.show()
```

```
# 4. Distribution of 'hp' (horsepower) with a boxplot to identify outliers
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['hp'], color='lightgreen')
plt.title('Boxplot of Horsepower (hp)')
plt.ylabel('Horsepower')
plt.show()
```

```
# 5. Frequency distribution of cars with different numbers of gears
plt.figure(figsize=(8, 6))
sns.countplot(x='gear', data=df, palette='coolwarm')
plt.title('Count of Cars by Number of Gears')
plt.xlabel('Number of Gears')
plt.ylabel('Count')
plt.show()
```

```
# 6. Relationship between horsepower (hp) and miles per gallon (mpg)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='hp', y='mpg', data=df, hue='cyl', palette='coolwarm', size='wt', sizes=(40, 400), alpha=0.7)
plt.title('Relationship between Horsepower and Miles per Gallon')
plt.xlabel('Horsepower (hp)')
plt.ylabel('Miles per Gallon (mpg)')
plt.legend(title='Number of Cylinders', loc='upper right')
plt.show()
```

```
# 7. Weight of cars (wt) based on cylinder type (cyl)
plt.figure(figsize=(8, 6))
sns.boxplot(x='cyl', y='wt', data=df, palette='Set2')
plt.title('Weight of Cars by Cylinder Type')
plt.xlabel('Number of Cylinders')
plt.ylabel('Weight (1000 lbs)')
plt.show()
```

```
# 8. Distribution of qsec (1/4 mile time) by transmission type (am)
plt.figure(figsize=(8, 6))
sns.histplot(df[df['am'] == 0]['qsec'], kde=True, color='red', label='Automatic', bins=10)
sns.histplot(df[df['am'] == 1]['qsec'], kde=True, color='blue', label='Manual', bins=10)
plt.title('Distribution of 1/4 Mile Time (qsec) by Transmission Type')
plt.xlabel('1/4 Mile Time (qsec)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
